```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
appstore_df = pd.read_csv('data/17k_appstore_games.csv')
appstore_df
```

```python
#Drop columns: in-app purchases, subtitle, icon URL, genres
appstore_dropcol_df = appstore_df.drop(columns=['Subtitle','Icon URL','In-app Purchases','Genres'])
appstore_dropcol_df
```

```python
#Drop rows with missing values
appstore_dropna_df = appstore_dropcol_df.dropna()
appstore_dropna_df
```

```python
#Check data types
appstore_dropna_df.dtypes
```

```python
#Convert dates to datetime format
appstore_dropna_df['Original Release Date'] = pd.to_datetime(appstore_dropna_df['Original Release Date'], format="%d/%m/%Y")
appstore_dropna_df
```

```python
appstore_dropna_df['Current Version Release Date'] = pd.to_datetime(appstore_dropna_df['Current Version Release Date'], format="%d/%m/%Y")
appstore_dropna_df
```

```python
#Convert datetime formats to %m/%d/%Y
appstore_dropna_df['Original Release Date'] = appstore_dropna_df['Original Release Date'].dt.strftime('%m/%d/%Y')
appstore_dropna_df['Current Version Release Date'] = appstore_dropna_df['Current Version Release Date'].dt.strftime('%m/%d/%Y')
appstore_dropna_df
```

```python
def str_to_num(x):
    return(len(x.replace(' ','').split(',')))
```

```python
appstore_almostcleaned_df=appstore_dropna_df
appstore_almostcleaned_df['Languages']= appstore_almostcleaned_df['Languages'].apply(str_to_num)
appstore_almostcleaned_df
```
